```python
# !pip3 install matplotlib pandas
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('~/Desktop/LSMTProject/results/mongodb/sar_mongodb-readAndModify.csv')
df['Workload'] = "mongodb-readAndModify"
df.columns
```

```python
useless = ["# hostname", 'interval', 'timestamp',"Workload"]
workload = df["Workload"][0]
for i in df.columns:
    if(i not in useless):
        plt.plot(df[i], label=i)
        plt.legend()
        plt.savefig(workload + "-sar-" + i.replace("/", "-") + ".png")
        plt.show()
```
